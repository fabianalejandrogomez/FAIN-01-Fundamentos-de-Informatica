```python
def ingreso():
    vta = int(input("ingrese el numero de venta de 1 a 3"))
    while vta <1 or vta >3:
        print("ingreso un numero incorrecto de venta")
        vta = int(input("ingrese el numero de venta de 1 a 3"))
    return vta

def cal_vta():
    qvta = 10
    vtasum = 0
    while qvta > 0:
        vta_imp = int(input("ingrese los importes de la venta"))
        vtasum = vtasum + vta_imp
        qvta = qvta - 1
    return vtasum

#programa principal
vta_a = ingreso()
vtasum_a = cal_vta()
print("la venta total de", vta_a,"fue:",vtasum_a )

vta_b = ingreso()
while vta_b == vta_a:
    print("elija otro numero de venta")
    vta_b = ingreso()
vtasum_b = cal_vta()
print("la venta total de", vta_b,"fue:",vtasum_b )

vta_c = ingreso()
while vta_c == vta_a or vta_c == vta_b:
    print("elija otro numero de venta")
    vta_c = ingreso()
vtasum_c = cal_vta()
print("la venta total de", vta_c,"fue:",vtasum_c )   

print("la venta total de todas las ventas fue", vtasum_a + vtasum_b + vtasum_c )   
    
```

```python
def factorial(x):
    n = x - 1
    prod = x
    while n > 0:
        prod = prod * n
        n = n - 1
        
    return prod
        
print(factorial(5))
```
